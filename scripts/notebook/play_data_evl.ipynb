```python
import os
import tensorflow as tf
import pandas as pd
import numpy as np
chunksize=10000
```

```python
evl = pd.read_csv("../../data/test.csv", iterator=True,chunksize=chunksize,na_values={'null','NaN'},keep_default_na=True)
```

```python
writer_evl = tf.python_io.TFRecordWriter("../../data/evl.tfrecords")
```

```python
d = None
counter = 0
row_counter = 0
for chunk in evl:
    counter += 1
    print ("Process %d, chunk" % counter)
    chunk["srch_ci"] = pd.to_datetime(chunk["srch_ci"],errors="coerce")
    chunk["srch_co"] = pd.to_datetime(chunk["srch_co"],errors="coerce")
    
    chunk["date_time"] = pd.to_datetime(chunk["date_time"],errors="coerce")

    chunk["srch_ci_month"] = chunk["srch_ci"].dt.month
    chunk["srch_ci_day"] = chunk["srch_ci"].dt.day

    chunk["srch_co_month"] = chunk["srch_co"].dt.month
    chunk["srch_co_day"] = chunk["srch_co"].dt.day
    
    chunk["month"] = chunk["date_time"].dt.month
    chunk["year"] = chunk["date_time"].dt.year

    chunk = chunk.drop(['date_time','srch_co','srch_ci'],axis=1)
    chunk = chunk.drop(['year'],axis=1)
    chunk = chunk.drop(['id'],axis=1)
    
    # Re-order
    cols = chunk.columns.tolist()
    cols = cols[-5:]+cols[:-5]
    chunk = chunk[cols]
    
    chunk["is_booking"] = 1
    chunk["cnt"] = 1

    cols = chunk.columns.tolist()
    cols = cols[:20]+cols[-2:]+cols[20:-2]
    chunk = chunk[cols]
    
    chunk = chunk.fillna(0)

    # Write to TFRecords
    for row in chunk.values:
        row_counter+=1
        example = tf.train.Example(features=tf.train.Features(
            feature={
                "label":tf.train.Feature(float_list=tf.train.FloatList(value=[row[-1]] )),
                "feature":tf.train.Feature(float_list=tf.train.FloatList(value=row[:] ))
            }
        ))
        writer_evl.write(example.SerializeToString())
print "Done"
```
